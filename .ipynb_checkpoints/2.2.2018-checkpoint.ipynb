Roll two dice separately.
- Find the PMF of the expirament that records the sum of two die rolls

```r
cat('First die:', X1, '\n')
X1 <- seq(1,6)
cat('Second die:', X1)
X2 <- seq(1,6)
```

```r
S <- expand.grid(X1=1:6, X2=1:6)
S
```

```r
(table(S$X1 + S$X2))/36
```

```r
barplot((table(S$X1 + S$X2))/36)
```
