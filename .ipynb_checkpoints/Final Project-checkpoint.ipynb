### d. Find precentage of days with AC within +/- 1 standard deviation from mean

```r
x_bar <- mean(Daily_2017$ACy)
s <- sd(Daily_2017$ACy)
lower_bound <- x_bar-s
upper_bound <- x_bar+s
counter <- 0
for (i in 1:length(Daily_2017$ACy)) {
    if (Daily_2017$ACy[i] >= lower_bound && Daily_2017$ACy[i] <= upper_bound) {
        counter <- counter + 1
    }
}
cat('Percentage daily AC within 1 sd:', counter/length(Daily_2017$ACy)*100)
```

### e. What percentage of days have PP > 35.4%

```r
counter <- 0
for (i in 1:length(Daily_2017$PPx)) {
    if (Daily_2017$PPx[i] >= 35.4) {
        counter <- counter + 1
    }
}
cat('Percentage of days with PP > 35.4:', counter/length(Daily_2017$ACy)*100)
```

### f. What percentage of days have PP > 35.4% and AC < 330

```r
counter <- 0
for (i in 1:length(Daily_2017$PPx)) {
    if (Daily_2017$PPx[i] >= 35.4 && Daily_2017$ACy[i] < 330) {
        counter <- counter + 1
    }
}
cat('Percentage of days with PP > 35.4 with AC < 330:', counter/length(Daily_2017$ACy)*100)
```

## 3 Assuming Daily AC follows a normal distribution estimate the mean using margin of error and C.I. of 95%

```r
n <- length(Daily_2017$ACy)
x_bar <- mean(Daily_2017$ACy)
s <- sd(Daily_2017$ACy)
lower_interval <- x_bar - 1.96*s/sqrt(n)
upper_interval <- x_bar + 1.96*s/sqrt(n)
cat('95% CI: (',lower_interval,',',upper_interval,')')
```
