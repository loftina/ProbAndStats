```r
n <- 10
x <- rep(0, 1000)
for (i in 1:length(x)){
    sd1 <- rt(n, df = 10)
    x[i] <- var(sd1)
}
hist(x)
```

```r
lower <- (4.85-(1.96*.75)/sqrt(20))
lower
upper <- (4.85+(1.96*.75)/sqrt(20))
upper
n <- 20
count <- 0
for (i in 1:1000){
    x <- rnorm(n, mean = 5, sd = .75)
    mu <- mean(x)
    if (mu > lower && mu < upper){
        count <- count + 1
    }
}
count/1000
```
