```r
count = 0
for (i in 1:100)
{
    y <- rep(1,100)
    for (j in 1:100)
    {
        y[j] <- rbinom(n = 1, size = 10, prob = .65)
    }
    print(y)
    z <- mean(y)
    print(z)
    w <- sd(y)
    print('w')
    print(w)
    print(z-1.96*w)
    print(z+1.96*w)
    if (z-1.96*w <= 6.5 && 6.5 <= z+1.96*w)
    {
        count <- count + 1   
    }
}
print(count)
```

```r
count = 0
for (i in 1:100)
{
    for (j in 1:100)
    {
        y[j] <- rbinom(n = 1, size = 10, prob = .65)
    }
    z <- mean(y)
    w <- sd(y)
    if (z-w <= 6.5-1.96*(1.6) && 6.5+1.96*(1.6) <= z+w)
    {
        count <- count + 1
    }
}
print(count)
```
