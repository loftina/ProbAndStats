## 3.

```r
x <- c(rep(0,190), rep(1,160), rep(2,150))
```

### a)

```r
y <- sample(x, 100)
cat("Sample 1 :", y, "\n")
table <- t(data.frame(table(y)/100))
for ( i in 1:4){
    y <- sample(x, 100)
    cat("Sample", i+1, ":", y, "\n")
    table <- rbind(table, t(data.frame(table(y)/100))["Freq",])
}
table
```

```r
row_to_keep = c(FALSE, TRUE, TRUE, TRUE, TRUE, TRUE)
table = table[row_to_keep,]
```

```r
cat("The sample portion with 0 scratches: ", mean(as.numeric(table[,1])), "\n")
cat("The sample portion with 1 scratches: ", mean(as.numeric(table[,2])), "\n")
cat("The sample portion with 2 scratches: ", mean(as.numeric(table[,3])), "\n")
```

### b)

```r
y = sample(x, 100)
cat("Sample 1 :", y, "\n")

table <- t(data.frame(mean(y), var(y), sd(y)))

for (i in 1:4){
    y <- sample(x, 100)
    cat("Sample ", i+1, " :", y, "\n")
    table <- cbind(table, t(data.frame(mean(y), var(y), sd(y))))
}
table
```
